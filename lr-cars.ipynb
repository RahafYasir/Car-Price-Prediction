```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/car-data/cardata.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns = ['Car_Name'])
df.head()
```

```python
catagorical_feature = [col for col in df.columns if df[col].dtype == 'O']
catagorical_feature 
```

```python
numerical_feature = [col for col in df.columns if df[col].dtype != 'O']
numerical_feature 
```

```python
df.isnull().sum()
```

```python
#convert catagorical data into numerical data
#now we have new columns for each type of catagorical data like Fuel_Type_CNG, Fuel_Type_Diesel, Fuel_Type_Petrol
df = pd.get_dummies(df)
df.head()
```

```python
features = ["Year", "Present_Price", "Kms_Driven", "Owner", "Fuel_Type_CNG", "Fuel_Type_Diesel", "Fuel_Type_Petrol", "Seller_Type_Dealer", "Seller_Type_Individual", "Transmission_Automatic", "Transmission_Manual"]
x = df[features]
y = df.Selling_Price
```

```python
x.head()
```

```python
y.head()
```

# Divid dataset into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=5)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
prediction = lr.predict(x_test)
```

```python
lr.score(x_train, y_train)
```

```python
plt.scatter(y_test, prediction)
```

# Evaluate model

```python
from sklearn import metrics

print('MAE: ',metrics.mean_absolute_error(y_test, prediction))
print('MSE: ',metrics.mean_squared_error(y_test, prediction))
print('RMSE: ',np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
